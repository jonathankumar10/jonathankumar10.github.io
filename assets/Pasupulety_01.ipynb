```python
import pandas as pd
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import math 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
%matplotlib inline
```

# Generate 20 data pairs (X, Y) using y = sin(2*pi*X) + N


### Use uniform distribution between 0 and 1 for X

### Sample N from the normal gaussian distribution


### Use 10 for train and 10 for test

```python
pi = np.pi
x = np.random.uniform(0,1,20)
n = np.random.normal(0,1)
```

```python
x.sort()
x
```

```python
n
```

```python
y = (np.sin(2*pi*x) + n)
```

```python
y
```

```python
plt.scatter(x, y)
plt.plot(x,y, color = 'r')
plt.grid(True)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.5)
```

Conversion of nd array to lists in Python

```python
X_train_list = X_train.tolist()
X_test_list = X_test.tolist()
y_train_list = y_train.tolist()
y_test_list = y_test.tolist()
```

Conversion of lists to dictionary

```python
dictionary1 = {}
for key in X_train_list: 
    for value in y_train_list: 
        dictionary1[key] = value 
        y_train_list.remove(value) 
        break
        
print(dictionary1)
```

```python
dictionary2 = {}
for key in X_test_list: 
    for value in y_test_list: 
        dictionary2[key] = value 
        y_test_list.remove(value) 
        break
        
print(dictionary2)
```

Creating sorted lists

```python
sort_orders1 = sorted(dictionary1.items())
sort_orders1
```

```python
sort_orders2 = sorted(dictionary2.items())
sort_orders2
```

```python
sorted_x = []
sorted_y = []
for i in sort_orders1:
    sorted_x.append(i[0])
    sorted_y.append(i[1])
```

```python
sorted_x2 = []
sorted_y2 = []
for i in sort_orders2:
    sorted_x2.append(i[0])
    sorted_y2.append(i[1])
```

Conversion of lists to Arrays

```python
X_train_sorted = np.array(sorted_x)
Y_train_sorted = np.array(sorted_y)
X_test_sorted = np.array(sorted_x2)
Y_test_sorted = np.array(sorted_y2)
```

Created a Graph with X_tain and Y_Train

```python
plt.scatter(X_train_sorted, Y_train_sorted, color = 'b',label = 'Train Data')
plt.plot(x,y, color= 'g', label = 'Total Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter(X_test_sorted, Y_test_sorted, color = 'b',label = 'Test Data')
plt.plot(x,y, color= 'g', label = 'Total Data')
plt.legend()
plt.grid(True)
plt.show()
```

# Using root mean square error, find weights of polynomial regression for order is 0, 1, 3, 9

Creating weights

```python
w=dict()
degrees = [0,1,3,9]
for degree in (degrees):
    w[degree] = (np.polyfit(X_train_sorted,Y_train_sorted,degree)).tolist()

print(w)
```

# Display weights in table

Display weights in tabular format

```python
df = pd.DataFrame.from_dict(w, orient='index').T
df = df.fillna(value= 0)
df = df.rename(columns ={0: 'M=0', 1 :'M=1', 3: 'M=3', 9:'M=9'})
df = df.rename({0 :'w0', 1:'w1', 2:'w2', 3:'w3', 4:'w4', 5:'w5', 6:'w6', 7:'w7', 8:'w8', 9:'w9'})
df
```

# Draw train error vs test error

```python
train_error = np.empty(10)
test_error = np.empty(10)
```

```python
for degree in range(10):
    est = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    est.fit(X_train_sorted.reshape(-1,1), Y_train_sorted)
    train_error[degree] = np.sqrt(mean_squared_error(Y_train_sorted, est.predict(X_train_sorted.reshape(-1,1))))
    test_error[degree] = np.sqrt(mean_squared_error(Y_train_sorted, est.predict(X_test_sorted.reshape(-1,1))))
```

```python
train_error
```

```python
test_error
```

```python
plt.plot(np.arange(10), train_error, 'bo-',color='blue', label='Train error')
plt.plot(np.arange(10), test_error,'ro-', color='red', label='Test error')
plt.ylim((0, 1))
plt.ylabel('Erms')
plt.xlabel('M')
plt.legend()
plt.grid(True)
plt.show()
```

# Now draw test and train error according to lamda

```python
lamda_train_error = np.empty(10)
lamda_test_error = np.empty(10)
```

```python
for degree in range(10):
    est = make_pipeline(PolynomialFeatures(degree), Ridge(degree))
    est.fit(X_train_sorted.reshape(-1,1), Y_train_sorted)
    lamda_train_error[degree] = np.sqrt(mean_squared_error(Y_train_sorted, est.predict(X_train_sorted.reshape(-1,1))))
    lamda_test_error[degree] = np.sqrt(mean_squared_error(Y_test_sorted, est.predict(X_test_sorted.reshape(-1,1))))
```
